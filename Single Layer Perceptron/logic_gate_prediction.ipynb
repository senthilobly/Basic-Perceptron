```python
import numpy as np
import matplotlib.pyplot as plt

# AND gate inputs and labels
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
w = np.random.randn(2)
b = 0.0

# Step activation function
def step(z):
    return 1 if z > 0 else 0

# Prediction
def predict(x):
    z = np.dot(w, x) + b
    return step(z)

# Training with true vs predicted plot
def train(X, y, lr=0.1, epochs=20):
    global w, b
    for epoch in range(epochs):
        y_preds = []
        for i in range(len(X)):
            y_pred = predict(X[i])
            y_preds.append(y_pred)
            error = y[i] - y_pred
            w += lr * error * X[i]
            b += lr * error
        print(f"True:      {y}")
        print(f"Predicted: {y_preds}")

        # Simple line plot: true vs predicted
        plt.plot(range(len(y)), y, 'bo-', label="True")
        plt.plot(range(len(y_preds)), y_preds, 'r.--', label="Predicted")
        plt.title(f"Epoch {epoch+1} Prediction")
        plt.xlabel("Sample Index")
        plt.ylabel("Output")
        plt.xticks(range(len(y)))
        plt.ylim(-0.5, 1.5)
        plt.grid(True)
        plt.legend()
        plt.show()

# Run training
train(X, y)
```
